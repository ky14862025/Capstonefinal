```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from pathlib import Path
import logging
import gc
from tqdm import tqdm
import re
from datetime import timedelta

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('employee_analysis.log'),
        logging.StreamHandler()
    ]
)

class EmployeeBehaviorAnalyzer:
    def __init__(self, base_path='/content/drive/MyDrive'):
        self.base_path = Path(base_path)
        self.scaler = MinMaxScaler()

        # Enhanced configuration with all detection parameters
        self.dataset_config = {
            'psychometric': {
                'file': 'psychometric.csv',
                'rename': {'user_id': 'user'},
                'dtype': {'user': 'category', 'employee_name': 'category'}
            },
            'logon': {
                'file': 'logon.csv',
                'parse_dates': ['date'],
                'dtype': {'user': 'category', 'pc': 'category', 'activity': 'category'}
            },
            'device': {
                'file': 'device.csv',
                'parse_dates': ['date'],
                'dtype': {'user': 'category', 'pc': 'category', 'activity': 'category'},
                'connection_terms': ['connect', 'insert', 'plug', 'attach'],
                'disconnection_terms': ['disconnect', 'remove', 'unplug', 'eject']
            },
            'email': {
                'file': 'email.csv',
                'parse_dates': ['date'],
                'dtype': {
                    'user': 'category',
                    'pc': 'category',
                    'to': 'string',
                    'cc': 'string',
                    'bcc': 'string',
                    'from': 'string'
                }
            },
            'file': {
                'file': 'file.csv',
                'parse_dates': ['date'],
                'dtype': {'user': 'category', 'pc': 'category', 'filename': 'string'},
                'sensitive_terms': [
                    'confidential', 'secret', 'private', 'classified',
                    'restricted', 'internal', 'proprietary', 'sensitive',
                    'privileged', 'top[ _-]secret', 'eyes[ _-]only'
                ],
                'risky_extensions': ['pem', 'key', 'bak', 'tmp', 'swp', 'cfg']
            }
        }

    def _load_dataset(self, dataset_name):
        """Load and process dataset with chunked processing"""
        config = self.dataset_config[dataset_name]
        file_path = self.base_path / config['file']

        try:
            chunks = []
            for chunk in tqdm(
                pd.read_csv(
                    file_path,
                    dtype=config.get('dtype', {}),
                    parse_dates=config.get('parse_dates', []),
                    chunksize=100000
                ),
                desc=f"Loading {dataset_name}"
            ):
                if 'rename' in config:
                    chunk = chunk.rename(columns=config['rename'])

                processor = getattr(self, f'_process_{dataset_name}', None)
                if processor:
                    chunk = processor(chunk)

                chunks.append(chunk)

            return pd.concat(chunks, ignore_index=True)

        except Exception as e:
            logging.error(f"Error loading {dataset_name}: {str(e)}")
            return pd.DataFrame()

    def _process_psychometric(self, df):
        """Process psychometric data"""
        return df

    def _process_logon(self, df):
        """Process logon data"""
        df['logon_count'] = 1
        df['logoff_count'] = (df['activity'] == 'Logoff').astype('int8')
        return self._process_datetime(df)

    def _process_device(self, df):
        """Enhanced device activity processing"""
        df['activity'] = df['activity'].astype(str).str.lower().str.strip()

        connect_pattern = '|'.join(self.dataset_config['device']['connection_terms'])
        disconnect_pattern = '|'.join(self.dataset_config['device']['disconnection_terms'])

        df['usb_connect'] = df['activity'].str.contains(connect_pattern, regex=True).astype('int8')
        df['usb_disconnect'] = df['activity'].str.contains(disconnect_pattern, regex=True).astype('int8')
        df['pc_change'] = (df.groupby('user')['pc'].shift() != df['pc']).astype('int8')

        df = self._process_datetime(df)
        if 'date' in df.columns:
            df = df.sort_values(['user', 'date'])
            df['time_since_last'] = df.groupby('user')['date'].diff()

        return df

    def _process_email(self, df):
        """Process email data"""
        df['email_count'] = 1
        df['to_count'] = df['to'].str.count(';').fillna(0).astype('int16') + 1
        df['cc_count'] = df['cc'].str.count(';').fillna(0).astype('int16') + (~df['cc'].isna()).astype('int16')
        df['bcc_count'] = df['bcc'].str.count(';').fillna(0).astype('int16') + (~df['bcc'].isna()).astype('int16')

        company_domains = ['@dtaa.com', '@company.com']
        pattern = '|'.join([re.escape(d) for d in company_domains])
        df['external_email'] = (~df['to'].str.contains(pattern, na=False, regex=True)).astype('int8')

        return self._process_datetime(df)

    def _process_file(self, df):
        """Enhanced file sensitivity detection"""
        df['file_count'] = 1
        patterns = '|'.join(self.dataset_config['file']['sensitive_terms'])

        filename_flags = df['filename'].str.contains(patterns, case=False, regex=True, na=False)
        content_flags = df['content'].str.contains(patterns, case=False, regex=True, na=False)
        ext_flags = df['filename'].str.extract(r'\.([a-zA-Z0-9]+)$', expand=False).str.lower().isin(
            self.dataset_config['file']['risky_extensions'])

        df['sensitive_file'] = (filename_flags | content_flags | ext_flags).astype('int8')
        return self._process_datetime(df)

    def _process_datetime(self, df, date_col='date'):
        """Process datetime features"""
        if date_col in df.columns:
            df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
            df['hour'] = df[date_col].dt.hour.astype('int8')
            df['after_hours'] = ((df['hour'] < 9) | (df['hour'] >= 17)).astype('int8')
        return df

    def _aggregate_features(self, df, prefix=''):
        """Robust feature aggregation with timedelta handling"""
        if df.empty or 'user' not in df.columns:
            return pd.DataFrame()

        # Convert timedeltas to seconds for aggregation
        timedelta_cols = [col for col in df.columns if pd.api.types.is_timedelta64_dtype(df[col])]
        for col in timedelta_cols:
            df[col] = df[col].dt.total_seconds()

        agg_dict = {
            'usb_connect': ['sum', 'mean'],
            'usb_disconnect': ['sum', 'mean'],
            'pc_change': 'sum',
            'after_hours': 'mean',
            'file_count': 'sum',
            'sensitive_file': ['sum', 'mean'],
            'email_count': 'sum',
            'external_email': ['sum', 'mean'],
            'time_since_last': ['mean', 'min', 'count']
        }

        # Filter for existing columns
        agg_dict = {k: v for k, v in agg_dict.items() if k in df.columns}

        aggregated = df.groupby('user').agg(agg_dict)

        # Flatten multi-index columns
        aggregated.columns = [
            f"{prefix}{col[0]}_{col[1]}" if isinstance(col, tuple) else f"{prefix}{col}"
            for col in aggregated.columns
        ]

        # Derived metrics
        if f"{prefix}usb_connect_sum" in aggregated.columns:
            aggregated[f"{prefix}connection_imbalance"] = (
                aggregated[f"{prefix}usb_connect_sum"] -
                aggregated[f"{prefix}usb_disconnect_sum"]
            )

            if f"{prefix}time_since_last_count" in aggregated.columns:
                aggregated[f"{prefix}pc_volatility"] = (
                    aggregated[f"{prefix}pc_change_sum"] /
                    aggregated[f"{prefix}time_since_last_count"]
                ).fillna(0)

        return aggregated

    def analyze_employee_behavior(self):
        """Complete analysis pipeline"""
        try:
            # Load all datasets
            psychometric = self._load_dataset('psychometric')
            logon = self._load_dataset('logon')
            device = self._load_dataset('device')
            email = self._load_dataset('email')
            file = self._load_dataset('file')

            # Merge all features
            result = psychometric.set_index('user')
            for name, data in [('logon', logon), ('device', device), ('email', email), ('file', file)]:
                if not data.empty:
                    features = self._aggregate_features(data, prefix=f"{name}_")
                    result = result.join(features, how='left')

            # Fill zeros and convert time metrics to minutes
            behavioral_cols = [c for c in result.columns if c not in ['O', 'C', 'E', 'A', 'N', 'employee_name']]
            result[behavioral_cols] = result[behavioral_cols].fillna(0)

            # Convert time columns from seconds to minutes
            time_cols = result.filter(regex='time_since_last').columns
            result[time_cols] = result[time_cols] / 60

            # Normalize numeric features
            numeric_cols = [c for c in result.columns if pd.api.types.is_numeric_dtype(result[c])]
            result[numeric_cols] = self.scaler.fit_transform(result[numeric_cols])

            return result

        except Exception as e:
            logging.error(f"Analysis failed: {str(e)}")
            raise

if __name__ == "__main__":
    analyzer = EmployeeBehaviorAnalyzer()
    results = analyzer.analyze_employee_behavior()

    # Safe display of results
    numeric_results = results.select_dtypes(include=[np.number])
    print("Behavioral features summary:")
    print(numeric_results.filter(regex='device_|file_').mean().to_markdown())

    # Save results
    results.to_csv('employee_behavior_analysis.csv')
    print("\nResults saved to employee_behavior_analysis.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.manifold import TSNE

# ---------- Module 1: Employee Behavior Analyzer with Psychometrics and Interaction Features ----------

class EmployeeBehaviorAnalyzer:
    def __init__(self, data_file="employee_behavior_analysis.csv"):
        self.data_file = data_file  # Path to the actual CSV file

    def load_employee_data(self):
        df = pd.read_csv(self.data_file)
        # Ensure the employee ID column is properly set up if needed, here assuming it's the 'user' column.
        df = df.rename(columns={'user': 'employee_id'})
        return df

    def analyze_employee_behavior(self):
        df = self.load_employee_data()

        # Group data by employee and aggregate as needed
        behavior_data = df.groupby('employee_id').agg({
            'O': 'mean',  # Openness
            'C': 'mean',  # Conscientiousness
            'E': 'mean',  # Extraversion
            'A': 'mean',  # Agreeableness
            'N': 'mean',  # Neuroticism
            'device_usb_connect_sum': 'sum',
            'device_usb_connect_mean': 'mean',
            'device_usb_disconnect_sum': 'sum',
            'device_connection_imbalance': 'mean',
            'device_pc_volatility': 'mean',
            'email_after_hours_mean': 'mean',
            'email_email_count_sum': 'sum',
            'email_external_email_sum': 'sum',
            'email_external_email_mean': 'mean',
            'file_after_hours_mean': 'mean',
            'file_file_count_sum': 'sum',
            'file_sensitive_file_sum': 'sum',
            'file_sensitive_file_mean': 'mean'
        }).reset_index()

        # Derived metrics for per-day averages or other needed metrics
        days_tracked = 30
        behavior_data['avg_device_usb_connections_per_day'] = behavior_data['device_usb_connect_sum'] / days_tracked
        behavior_data['avg_email_external_email_per_day'] = behavior_data['email_external_email_sum'] / days_tracked
        behavior_data['avg_file_sensitive_file_per_day'] = behavior_data['file_sensitive_file_sum'] / days_tracked

        # 🔥 Add interaction features
        behavior_data['usb_x_openness'] = behavior_data['device_usb_connect_sum'] * behavior_data['O']
        behavior_data['email_x_conscientiousness'] = behavior_data['email_email_count_sum'] * behavior_data['C']
        behavior_data['file_x_neuroticism'] = behavior_data['file_file_count_sum'] * behavior_data['N']

        return behavior_data


# ---------- Module 2: Insider Threat Detector with Clustering Tuning and Visuals ----------

class InsiderThreatDetector:
    def __init__(self):
        self.scaler = StandardScaler()
        self.pca = PCA(n_components=2)
        self.anomaly_model = IsolationForest(contamination=0.1, random_state=42)

    def find_optimal_k(self, data, max_k=10):
        distortions = []
        for k in range(2, max_k + 1):
            km = KMeans(n_clusters=k, random_state=42)
            km.fit(data)
            distortions.append(km.inertia_)

        plt.figure(figsize=(8, 5))
        plt.plot(range(2, max_k + 1), distortions, marker='o')
        plt.title("Elbow Method for Optimal k")
        plt.xlabel("Number of clusters")
        plt.ylabel("Inertia")
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    def detect_threats(self, behavior_data, use_psychometrics=True):
        # Feature set
        if use_psychometrics:
            features = behavior_data[[
                'avg_device_usb_connections_per_day', 'avg_email_external_email_per_day', 'avg_file_sensitive_file_per_day',
                'O', 'C', 'E', 'A', 'N', 'usb_x_openness', 'email_x_conscientiousness', 'file_x_neuroticism'
            ]]
        else:
            features = behavior_data[[
                'avg_device_usb_connections_per_day', 'avg_email_external_email_per_day', 'avg_file_sensitive_file_per_day'
            ]]

        scaled_features = self.scaler.fit_transform(features)
        reduced_features = self.pca.fit_transform(scaled_features)

        # Elbow check for KMeans clustering
        self.find_optimal_k(scaled_features)

        # Clustering
        cluster_model = KMeans(n_clusters=3, random_state=42)
        clusters = cluster_model.fit_predict(reduced_features)
        behavior_data['cluster'] = clusters

        # Clustering metrics
        silhouette = silhouette_score(reduced_features, clusters)
        calinski = calinski_harabasz_score(reduced_features, clusters)
        davies = davies_bouldin_score(reduced_features, clusters)

        print("\n📈 Clustering Quality Metrics:")
        print(f"Silhouette Score: {silhouette:.2f}")
        print(f"Calinski-Harabasz Score: {calinski:.2f}")
        print(f"Davies-Bouldin Score: {davies:.2f}")

        # Anomaly detection
        anomaly_scores = self.anomaly_model.fit_predict(scaled_features)
        behavior_data['threat_flag'] = anomaly_scores
        behavior_data['anomaly_score'] = self.anomaly_model.decision_function(scaled_features)

        # Visualize clusters and threats
        self.visualize_clusters(behavior_data, reduced_features)

        return behavior_data[behavior_data['threat_flag'] == -1]

    def visualize_clusters(self, data, reduced_features):
        print("\n🖼️ Visualizing PCA & Clusters...")
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=reduced_features[:, 0], y=reduced_features[:, 1],
                        hue=data['cluster'], style=data['threat_flag'],
                        palette='deep', s=100)
        plt.title("PCA View of Clusters and Threat Flags")
        plt.xlabel("PCA 1")
        plt.ylabel("PCA 2")
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # Alternative visualization using t-SNE
        tsne = TSNE(n_components=2, random_state=42)
        tsne_features = tsne.fit_transform(data[['avg_device_usb_connections_per_day', 'avg_email_external_email_per_day', 'avg_file_sensitive_file_per_day']])
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=tsne_features[:, 0], y=tsne_features[:, 1], hue=data['cluster'], style=data['threat_flag'], palette='deep', s=100)
        plt.title("t-SNE View of Clusters and Threat Flags")
        plt.xlabel("t-SNE 1")
        plt.ylabel("t-SNE 2")
        plt.grid(True)
        plt.tight_layout()
        plt.show()


# ---------- 🧪 Run & Compare ----------

if __name__ == "__main__":
    print("🚀 Running Enhanced Insider Threat Detection System\n")

    # Initialize the behavior analyzer and detect threats
    analyzer = EmployeeBehaviorAnalyzer(data_file="employee_behavior_analysis.csv")
    behavior_df = analyzer.analyze_employee_behavior()

    detector = InsiderThreatDetector()

    print("➡️ With Psychometrics + Interactions:")
    threats_psy = detector.detect_threats(behavior_df, use_psychometrics=True)

    print("\n➡️ Without Psychometrics:")
    threats_behavior_only = detector.detect_threats(behavior_df, use_psychometrics=False)

    print(f"\n🔐 Threats (with psychometrics): {len(threats_psy)}")
    print(f"🧍 Threats (behavior only): {len(threats_behavior_only)}")

    # 📊 Top 10 most anomalous employees (with psychometrics)
    print("\n💣 Top 10 Highest-Risk Employees (Without Psychometrics):")
    top_threats = threats_psy.sort_values(by='anomaly_score').head(10)
    print(top_threats[['employee_id', 'anomaly_score', 'cluster',
                       'device_usb_connect_sum', 'device_usb_connect_mean', 'device_usb_disconnect_sum',
                       'email_after_hours_mean', 'file_file_count_sum']].to_string(index=False))
```

```python
import pandas as pd
import numpy as np
import logging
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Configure logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

class AdvancedInsiderThreatDetector:
    def __init__(self, psych_weight=0.4, behavior_weight=0.6):
        self.psych_weight = psych_weight
        self.behavior_weight = behavior_weight
        self.scaler = StandardScaler()
        self.anomaly_model = IsolationForest(contamination=0.1, random_state=42)
        self.pca = PCA(n_components=2, random_state=42)
        self.interaction_features = ['usb_x_openness', 'email_x_conscientiousness', 'file_x_neuroticism']

    def preprocess(self, df: pd.DataFrame) -> pd.DataFrame:
        # Create derived metrics
        df['avg_device_usb_connections_per_day'] = df['device_usb_connect_sum'] / 30
        df['avg_email_external_email_per_day'] = df['email_external_email_sum'] / 30
        df['avg_file_sensitive_file_per_day'] = df['file_sensitive_file_sum'] / 30

        # Create interaction terms
        df['usb_x_openness'] = df['device_usb_connect_sum'] * df['O']
        df['email_x_conscientiousness'] = df['email_email_count_sum'] * df['C']
        df['file_x_neuroticism'] = df['file_file_count_sum'] * df['N']

        return df

    def prepare_features(self, df: pd.DataFrame, use_psych=True) -> pd.DataFrame:
        if use_psych:
            feature_cols = [
                'avg_device_usb_connections_per_day', 'avg_email_external_email_per_day',
                'avg_file_sensitive_file_per_day', 'O', 'C', 'E', 'A', 'N'
            ] + self.interaction_features
        else:
            feature_cols = [
                'avg_device_usb_connections_per_day', 'avg_email_external_email_per_day',
                'avg_file_sensitive_file_per_day'
            ]

        df = df.dropna(subset=feature_cols)
        features = self.scaler.fit_transform(df[feature_cols])
        return pd.DataFrame(features, columns=feature_cols, index=df.index)

    def find_optimal_k(self, data):
        distortions = []
        silhouette_scores = []

        k_range = range(2, min(len(data), 10))
        for k in k_range:
            km = KMeans(n_clusters=k, random_state=42).fit(data)
            distortions.append(km.inertia_)
            silhouette_scores.append(silhouette_score(data, km.labels_))

        optimal_k = k_range[np.argmax(silhouette_scores)]

        plt.plot(k_range, distortions, marker='o')
        plt.title("Elbow Method")
        plt.xlabel("k")
        plt.ylabel("Inertia")
        plt.grid(True)
        plt.show()

        return optimal_k

    def cluster_and_detect(self, features_df: pd.DataFrame) -> pd.DataFrame:
        reduced_features = self.pca.fit_transform(features_df)
        optimal_k = self.find_optimal_k(reduced_features)
        cluster_model = KMeans(n_clusters=optimal_k, random_state=42)
        clusters = cluster_model.fit_predict(reduced_features)

        features_df['cluster'] = clusters
        results = []

        for cluster_id in np.unique(clusters):
            cluster_data = features_df[features_df['cluster'] == cluster_id].drop(columns='cluster')
            anomaly_scores = self.anomaly_model.fit_predict(cluster_data)
            results.append(pd.DataFrame({
                'user': cluster_data.index,
                'cluster': cluster_id,
                'anomaly_score': self.anomaly_model.decision_function(cluster_data),
                'threat_flag': (anomaly_scores == -1).astype(int)
            }))

        return pd.concat(results)

    def generate_risk_report(self, anomalies_df: pd.DataFrame, original_df: pd.DataFrame) -> pd.DataFrame:
        report = []
        for _, row in anomalies_df[anomalies_df['threat_flag'] == 1].iterrows():
            user_id = str(row['user'])
            user_data = original_df.loc[user_id]

            psych_risks = []
            if user_data.get('N', 0) > 0.6:
                psych_risks.append('high neuroticism')
            if user_data.get('C', 0) < 0.4:
                psych_risks.append('low conscientiousness')

            behavior_risks = []
            if user_data.get('device_connection_imbalance', 0) > 0.7:
                behavior_risks.append("abnormal device connections")
            if user_data.get('file_sensitive_file_mean', 0) > 0.6:
                behavior_risks.append("excessive sensitive file access")
            if user_data.get('email_external_email_mean', 0) > 0.5:
                behavior_risks.append("frequent external communications")

            combined_risk = (1 + len(psych_risks) + len(behavior_risks)) * abs(row['anomaly_score'])

            report.append({
                'user': user_id,
                'cluster': row['cluster'],
                'anomaly_score': row['anomaly_score'],
                'psych_risks': ", ".join(psych_risks) if psych_risks else "None",
                'behavior_risks': ", ".join(behavior_risks) if behavior_risks else "None",
                'combined_risk': combined_risk
            })

        return pd.DataFrame(report).sort_values(by="combined_risk", ascending=False)

    def visualize(self, features_df, anomalies_df):
        pca = PCA(n_components=2, random_state=42)
        vis_pca = pca.fit_transform(features_df.drop(columns='cluster'))
        tsne = TSNE(n_components=2, random_state=42)
        vis_tsne = tsne.fit_transform(features_df.drop(columns='cluster'))

        # PCA View
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        sns.scatterplot(x=vis_pca[:, 0], y=vis_pca[:, 1], hue=features_df['cluster'], palette='Set2')
        plt.title("PCA Clusters")

        # t-SNE View
        plt.subplot(1, 2, 2)
        sns.scatterplot(x=vis_tsne[:, 0], y=vis_tsne[:, 1], hue=anomalies_df['threat_flag'], palette={0: 'grey', 1: 'red'})
        plt.title("t-SNE Threat Visualization")
        plt.tight_layout()
        plt.show()
```

```python
if __name__ == "__main__":
    df = pd.read_csv("employee_behavior_analysis.csv", index_col='user')
    detector = AdvancedInsiderThreatDetector()

    df = detector.preprocess(df)
    features = detector.prepare_features(df, use_psych=True)
    anomalies = detector.cluster_and_detect(features)
    report = detector.generate_risk_report(anomalies, df)

    print("🚨 Top Threats:")
    print(report.head(10).to_markdown())

    detector.visualize(features, anomalies)
```
